```python
import requests
import json
from requests.adapters import HTTPAdapter
from requests.packages.urllib3.util.retry import Retry
import timeit as tt
```

```python
#sometimes M1 site doesn't let you access stuff if you request too often
#this is for retrying in case they don't give us access
session = requests.Session()
retry = Retry(connect=3, backoff_factor=0.5)
adapter = HTTPAdapter(max_retries=retry)
session.mount('http://', adapter)
session.mount('https://', adapter)
```

```python
#MAX refers to the number of entrants. For now it is hard coded. 
#Update for the future: make MAX and currentYear not hard coded
MAX = 14843
currentYear = 2020
allData = {}
```

```python
#access the API of M-1gp.com to get the data of all entrants

start = tt.default_timer()     
for groupId in range(MAX):
    try:
        groupId +=1
        r = session.get('https://www.m-1gp.com/combi/json/{0}.json'.format(groupId))
        combi = r.json()
        tempDict = {}
        #gets everything that seems important
        tempDict['name'] = combi['combiname']
        tempDict['nameKana'] = combi['combikana']
        #not sure what to do with the image but got it anyways
        tempDict['image'] = combi['image']
        tempDict['formYear'] = int(combi['orga_year'])
        tempDict['status'] = combi['belong']
        #members is currently useless right now
        tempDict['members'] = combi['member']
        tempDict['numberOfMembers'] = len(combi['member'])
        yearResultLst = []
        for key in range(2015, 2020):
            numKey = 0
            if str(key) in combi['entrant']:
                while str(numKey) in combi['entrant'][str(key)]:
                    numKey +=1
                numKey -=1
                yearResultLst.append((combi['entrant'][str(key)][str(numKey)]['war'], combi['entrant'][str(key)][str(numKey)]['result']))
        tempDict['result'] = yearResultLst
        allData[groupId] = tempDict
    except:
        print(groupId)
with open('M1_Data.json', 'w') as f:
    json.dump(allData, f)       
stop = tt.default_timer() 
print('It took', stop-start, 'seconds')
```
